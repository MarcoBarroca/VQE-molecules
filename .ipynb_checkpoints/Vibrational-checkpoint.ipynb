```python
from qiskit.algorithms import VQE, NumPyMinimumEigensolver, NumPyEigensolver #Algorithms

#Qiskit odds and ends
from qiskit.circuit.library import EfficientSU2, TwoLocal
from qiskit.algorithms.optimizers import COBYLA, SPSA, SLSQP, L_BFGS_B
from qiskit.opflow import Z2Symmetries, X, Y, Z, I, PauliSumOp, Gradient
from qiskit import IBMQ, BasicAer, Aer
from qiskit.utils import QuantumInstance, algorithm_globals
from qiskit.utils.mitigation import CompleteMeasFitter #Measurement error mitigatioin
from qiskit.tools.visualization import circuit_drawer
from qiskit.providers.aer import QasmSimulator
from qiskit.providers.aer.noise import NoiseModel
from qiskit.test.mock import FakeManila, FakeNairobi, FakeGuadalupe

#qiskit_nature
from qiskit_nature.circuit.library import UVCCSD, CHC, VSCF
from qiskit_nature.drivers import UnitsType, Molecule
from qiskit_nature.mappers.second_quantization import ParityMapper, BravyiKitaevMapper, JordanWignerMapper, DirectMapper
from qiskit_nature.converters.second_quantization import QubitConverter
from qiskit_nature.drivers.second_quantization import GaussianForcesDriver
from qiskit_nature.problems.second_quantization import VibrationalStructureProblem
from qiskit_nature.algorithms.ground_state_solvers.minimum_eigensolver_factories import NumPyMinimumEigensolverFactory
from qiskit_nature.algorithms import AdaptVQE,VQEUCCFactory, GroundStateEigensolver, QEOM, NumPyEigensolverFactory, ExcitedStatesEigensolver

#Python odds and ends
import matplotlib
import matplotlib.pyplot as plt
import matplotlib
import pylab
import numpy as np
import os
import pyscf
from IPython.display import display, clear_output

from datetime import datetime
%matplotlib inline
%config InlineBackend.figure_format = 'svg' # Makes the images look nice

#IBMQ.load_account()
#provider = IBMQ.get_provider(group='open')
#set Backends
#simulators
backend_stv = Aer.get_backend('aer_simulator_statevector')
#Noisy simulator
backend_aer = Aer.get_backend('aer_simulator')
device_backend=FakeGuadalupe()
device = QasmSimulator.from_backend(device_backend)
noise_model = None
coupling_map = device.configuration().coupling_map
noise_model = NoiseModel.from_backend(device)
basis_gates = noise_model.basis_gates
#Real Devices
#backend_manil = provider.get_backend('ibmq_manila')
#backend_qsm_ibm=provider.get_backend('ibmq_qasm_simulator')

sh=1024 #shots

#solvers
npme = NumPyMinimumEigensolver()
npe=NumPyEigensolver()
```

```python
def exact_solver(problem, converter):
    solver = NumPyMinimumEigensolverFactory(use_default_filter_criterion=True)
    calc = GroundStateEigensolver(converter, solver)
    result = calc.solve(problem)
    return result
```

```python
seed = 62
algorithm_globals.random_seed = seed
#qi=backend_stv
qi= QuantumInstance(backend_aer, seed_simulator=seed, seed_transpiler=seed, #Noisy Quantum Instance
                          coupling_map=coupling_map,noise_model=noise_model,
                          measurement_error_mitigation_cls=CompleteMeasFitter)
optimizer = COBYLA(maxiter=50)
mapper=DirectMapper()
backend=backend_stv
num_modals=[2,2,2,2]
init_state = VSCF(num_modals=num_modals)
```

```python
# if you ran Gaussian elsewhere and already have the output file
driver = GaussianForcesDriver(logfile="aux_files/CO2_freq_B3LYP_631g.log")
```

```python
problem = VibrationalStructureProblem(driver, num_modals=num_modals, truncation_order=2)
second_q_ops = problem.second_q_ops()
converter = QubitConverter(mapper=DirectMapper())
qubit_op = converter.convert(second_q_ops[0])
sym=Z2Symmetries.find_Z2_symmetries(qubit_op)
converter = QubitConverter(mapper=DirectMapper(),)
```

```python
print(qubit_op.num_qubits)
init_state.num_qubits
```

```python
backend=backend_stv
#ansatz
ansatze = [UVCCSD(converter,num_modals=num_modals,initial_state = init_state)
           ,CHC(num_qubits=qubit_op.num_qubits,excitations=[[(-2,-1),(0,-3)]],reps=1,ladder=True,initial_state=init_state)
           ,TwoLocal(num_qubits=qubit_op.num_qubits,rotation_blocks=['ry','rz'] ,entanglement_blocks='cx',entanglement='linear',reps=3,initial_state=init_state,name="RYRZ")
           #,EfficientSU2(num_qubits=qubit_op.num_qubits, su2_gates=['swap','rz'],entanglement='linear',reps=2,initial_state=init_state,name="SWAPRZ")
          ]

begin = datetime.now()
results={}

for i, ansatz in enumerate(ansatze):
    
    beginl=datetime.now()
    results[type(ansatz).__name__]={}
    def callback(eval_count, parameters, mean, std):  
        # Overwrites the same line when printing
        display("Evaluation: {}, Energy: {}, Std: {}".format(eval_count, mean, std))
        clear_output(wait=True)
        counts.append(eval_count)
        values.append(mean)
        params.append(parameters)
        deviation.append(std)

    counts = []
    values = []
    params = []
    deviation = []
    
    vqe = VQE(ansatz
              , optimizer
              ,quantum_instance=backend_aer
              ,callback=callback
              #,initial_point=initial_point
             )
    vqe_solver=GroundStateEigensolver(converter,vqe)
    vqe_result=vqe_solver.solve(problem)
    #results[type(ansatz).__name__]['opt_pt']=vqe_result.optimal_point
    results[type(ansatz).__name__]['eigenvalues']=vqe_result.computed_vibrational_energies
    results[type(ansatz).__name__]['results']=vqe_result
    results[type(ansatz).__name__]['counts'] = np.asarray(counts)
    results[type(ansatz).__name__]['values'] = np.asarray(values)
        
    endl = datetime.now()
    results[type(ansatz).__name__]['Runtime'] = endl-beginl
    
end = datetime.now()
results['Total Runtime']=end-begin

#classic solver
result_exact = exact_solver(problem, converter = converter)
```

```python
#Print Results
for i,ansatz in enumerate(ansatze):
    print("VQE Results for",type(ansatz).__name__,":", results[type(ansatz).__name__]['eigenvalues'])
print(f'Exact Result: {result_exact.computed_vibrational_energies}')

print("\n")

print ("Total Runtime:",results['Total Runtime'])
for i,ansatz in enumerate(ansatze):
    print ("Runtime for ",type(ansatz).__name__,":",results[type(ansatz).__name__]['Runtime'])
    
#Plot Convergence
pylab.rcParams['figure.figsize'] = (12, 8)
for i,ansatz in enumerate(ansatze):
    pylab.plot(results[type(ansatz).__name__]['counts'],
               results[type(ansatz).__name__]['values'],
               label=type(ansatz).__name__)
pylab.axhline(y=np.real(result_exact.computed_vibrational_energies),label='Classical Solver',linestyle='--')
pylab.xlabel('Eval count')
pylab.ylabel('Energy')
pylab.title('CO2 groundstate convergence for different ansatze')
pylab.legend(loc='upper right');
```

```python
for i, ansatz in enumerate(ansatze):
    print('*******************VQE Result for ', type(ansatz).__name__,'*******************\n')

    print(problem.interpret(results[type(ansatz).__name__]['results']))

print('\n*******************Exact Result*******************\n')

print(problem.interpret(result_exact))
```
